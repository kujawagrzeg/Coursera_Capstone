Introduction 

The project is being realised for the stakeholder, who want to open a new coffee point - cafeteria in Moscow. It is supposed to be a place mainly aiming for peolpe buying coffee to go - little space for high income. For that reason we will bo looking for a place where many people are commuting, traveling every day for their way to work or home. Best suited will be the place close to the centre of the city or to the transpotation joint, simultaneously not close to places with a great density of restaurants and cafes. For that reason we will look around Moscow metro stations - where every day milions of people go back and forth on their ways to schools, universities, works, families and homes. 

Data

We will take into consideration metro stations from Moscow - the source of data for their exact positions on the map of Moscow will be the wiki page https://en.wikipedia.org/wiki/List_of_Moscow_Metro_stations. For better understanding and clear examination we will use English translations or transcriptions of the original, Russian names of the stations. The location (geopositioning) will allow us to determine the close neighbourghood of the stations, and then, analyse their surroundings, searching for exisiting restaurants and other attractions in the area. 

Methodology

We are taking into conisderation two factors - distance from the centre of Moscow (for that aim we assume it being at position 55°45′21″N 37°37′2″E, according also to wikipedia page https://en.wikipedia.org/wiki/Moscow ), and the number of places similar in character to the place we try to build - reataurants and cafes. The numbers will be taken on the basis of the Foursquare service queries in the general category of food. Then we find the areas (stations) having the least number of such places around (in 5 km radius); coming from the least number we analyse if those stations are close enough to the centre of the city (assuming, that our stakeholders prefer places not further than 10 km from the geographical centre of Moscow, letting it be a bit more distant than it for especially favouring locations). There, we do not use any mathematical formula there any more to find out the best location, and leave it to arbitrary decision, giving descriptive analysis for the best place we find fitting to our assumptions.

Results and discussion

The highest number of food venues (category 4d4b7105d754a06374d81259 in Foursquare) is cumulated around the geographical and historical centre of the city (clusters of areas up to 8 km from the centre of the city). Of course, there is limitation to this method - to find out the exact distances and character of such places, we will need a detailed analysis of certain stations, which can be also provided to the stakeholders on the further data analysis stage. There we should concentrate on the areas around stations in the circle around the centre, in diameter of 8-10 kms from the central point of the city.
