Our task is to present the stakeholder with a conception and localization of the coffee point he is willing to open. The place chosen by the stakeholder is Poznań, city in the western part of Poland. We are supposed to suggest the theme or dominant character of the place, favourably targeted to young people (25-40 years old) and approximate localization for searching the exact venue to rent or buy.<br>
Poznań is one of the biggest Polish cities, former capital city of Poland and, currently, the centre of the region and of the voivodeship (the formal unit of Polish administration). It is inhabited by around 500 thousand people, with the fourth of this number being students of the local universities. As the centre of the city area of Ponznań, it is a place of work for many people living in the towns and villages nearby, commuting every day to their workplaces and back.<br>
Poznań is also the local cultural centre and the point of the social life with a big number of restaurants, cafes, food points, etc. Because of that we need to think carefully about the placement of the new coffee place, so that it will guarantee enough clients for the place to generate profits. We should avoid places far from the central districts, yet be careful not to choose place already full of similar joints.<br>
Basing on the collected data (further description of all of the data and their analysis in the next part of the project) we decide that the leading topic, character of the place will be cycling - coffee shop for cyclist, visually themed also with design inspired by bicycles and cycling.<br>
Our next move is to predict the best location for the venue. We will do this basing on our strategy, publicly known data and the Foursquare tools.
